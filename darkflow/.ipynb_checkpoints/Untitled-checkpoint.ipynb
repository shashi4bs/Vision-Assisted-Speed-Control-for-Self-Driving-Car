```python
import cv2
from darkflow.net.build import TFNet
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
import tensorflow as tf
```

```python
config = tf.ConfigProto(log_device_placement=True)
config.gpu_options.allow_growth = True
with tf.Session(config=config) as sess:
    options = {
        'model': 'cfg/yolo.cfg',
        'load' : 'bin/yolo.weights',
        'threshold': 0.2,
        'gpu': 1
    }
    tfnet = TFNet(options)
```

```python
img = cv2.imread('./preview.png', cv2.IMREAD_COLOR)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#img = np.asarray(img,dtype=np.int32)
```

```python
result = tfnet.return_predict(img)
```

```python
result
```

```python
import numpy as np
max(img.ravel())
```

```python
plt.figure(figsize=(0,12))
for r in result:
    print(r)
    t1 = (r['topleft']['x'] if r['topleft']['x']<500 else 499,r['topleft']['y'] if r['topleft']['y']<250 else 248)
    br = (r['bottomright']['x'],r['bottomright']['y'])
    label = r['label']
    img = cv2.rectangle(img, t1, br,(0,255,0),5)
    img = cv2.putText(img, label, t1, cv2.FONT_ITALIC, 0.8, (255,255,255), 2)
    
plt.imshow(img)
plt.show()
```
